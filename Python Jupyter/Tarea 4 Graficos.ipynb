```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import numpy as np
import pandas as pd
import webbrowser 
import matplotlib 
import matplotlib.pyplot as plt
```

```python
UBER = pd.read_csv('UBER_Dataset.csv')
UBER.info()
```

```python
plt.style.use('seaborn')
colores = ['black']
UBER.Default_Prep_Time.hist(color=colores)
```

#Aca se observa que el tiempo de preparacion de las comida en los restaurantes es de alrededor de 20 a 30 minutos.

```python
plt.style.use('Solarize_Light2')
Tiempos = UBER.groupby('Day_Week')['Delivery_Request'].max()
colores = ['red','teal','lightblue','indigo','gold','coral','black']
Tiempos.plot.barh(color=colores)
```

#Aca se puede observar que los dias con mayo solicitud de entrega por parte del cliente son los fines de semana, teniendo un mayor porcentaje los domingos

```python
plt.style.use('seaborn-pastel')
Rapidez_Entrega = UBER.groupby('Courier')['Courier_Acceptance'].sum()
Rapidez_Entrega_max = Rapidez_Entrega.sort_values(ascending=False).head(5)
colores = ['red','teal','lightblue','indigo','gold','coral','black']

Rapidez_Entrega_max.plot(kind='bar', color=colores, title='TOP 5 DE MENSAJEROS CON MAYOR SOLICITUDES')

```

##Aca se observa que en el top 5 de los mensajeros, el rabajador con el ID D5A2C869 tiene un mayor numero de solicitudes aceptadas, mientras que los demas mantienen un rango parecido. 

```python
plt.style.use('seaborn-pastel')
Rapidez_Entrega = UBER.groupby('Courier')['Courier_Acceptance'].sum()
Rapidez_Entrega_max = Rapidez_Entrega.sort_values(ascending=True).head(5)
colores = ['Black','white','Black','white','Black','coral']

Rapidez_Entrega_max.plot(kind='bar', color=colores, title='TOP 5 DE MENSAJEROS CON MENOR SOLICITUDES')
```

## Aca se observa el caso contrario, el top 5 de los mensajeros que poseen menor solicitudes de pedidos aceptadas, dejando claro que 3 de ellos mantienen un rango bajo igual y constante.

```python
UBER.sort_values(by=['Delivery_Request','Day_Week']).head(5)
```

##Aca se analiza que el dia con menor solicitud de pedidos son los lunes.

```python
plt.style.use('seaborn-whitegrid')
Entrega_completa = UBER.groupby('Courier')['End_to_End_Time'].sum()
Entrega_completa_max = Entrega_completa.sort_values(ascending=False).head(5)
colores = ['Black','white','Green','white','Black']
Entrega_completa_max.plot(kind='bar', color=colores, title='TOP 5 DE MENSAJEROS CON End to End COMPLETO')
```

#Aca se observa que el mensajero con el ID 5B3DCBC0 tiene el mayor rating de end to end, esto quiere decir que es uno de los mensajero que ha logrado mayormente hacer todo el proceso completo desde la solicitud del cliente.
