# Aca se decide agregar una columna nueva con informacion de los barrios de la ciudad de Chicago para poder realizar un analisis mas especifico y exacto.

```python
plt.style.use('dark_background')
Total = Inspeccion.groupby('Riesgo')['Riesgo'].count()
Total_max = Total.sort_values(ascending=False)
colores = ['red','yellow','lightblue','green','gold','coral','black']
Total_max.plot(kind='bar', color=colores, title='Categoria de Riegos')
```

# Aca podemos observar alrededor de un 60% de los riesgos encontrados son de categoria alta, esto quiere decir que hay una gran mayoria de locales que no poseen las condiciones otimas para laborar o brindar un servicio.

```python
plt.style.use('classic')
inspecciones = Inspeccion.groupby('Tipo de Inspeccion')['Tipo de Inspeccion'].count()
inspecciones_max = inspecciones.sort_values(ascending=False).head(5)
colores = ['Black','purple','pink','violet','purple']
inspecciones_max.plot(kind='bar', color=colores, title='TOP 5 de las principales inspecciones')
```

## Aca se hace un analisis de los tipos de inspecciones como licencias, condiciones del lugar y quejas.El tipo de Inspeccion que mas se realiza son las de licencias, se puede observar que la mayoria de los locales tienen problemas con ellas y esto causa que muchos de los locales sean cerrados o clausurados temporalmente.

```python
import seaborn as sns
```

```python
sns.countplot(Inspeccion['Resultado'],palette='ocean')
```

# Aca se observa que la mayoria de locales han pasado las inspcciones pero con alguna condicion establecida.

```python
grafico=sns.countplot(x='Resultado',hue='Tipo de Local',palette='hot_r',data=Inspeccion)
grafico.set(title='Resultado por Local',xlabel='Resultado',ylabel='Medida')
plt.show()
```

# Aca podemos observar mas detallado los resultados de cada inspeccion por locales, donde se puede ver que existe una mayor cantidad de restaurantes que han pasado las inspacciones pero con condiciones, mientras que las paleterias, los catering y licoreras han pasado por completo las inspecciones.


```python
sns.countplot(Inspeccion['Barrio'],palette='icefire')
```

# Aca pdemos observar que los barrios  Washington Park y South Chicago son a los que mayormente se les han hecho algun tipo de inspeccion.Esto quiere decir que son los barrios con mayor locales abiertos y con mayor cantidad de situaciones.

```python
plt.style.use('grayscale')
Fecha = Inspeccion.groupby('Fecha Inspeccion')['Fecha Inspeccion'].count()
Fecha_max = Fecha.sort_values(ascending=False).head(4)
colores = ['Black','pink','purple','black','purple',"Pink"]
Fecha_max.plot(kind='bar', color=colores, title='Fecha de  Inspecciones')
```
